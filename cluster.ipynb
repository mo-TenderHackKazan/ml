```python
import pandas as pd
import numpy as np
import pickle

with open('dump.pickle', 'rb') as file:
    data = pickle.load(file)
```

```python
data
```

```python
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=0.1, min_samples=40, metric='cosine').fit(np.array(data['embeddings'].tolist()))
```

```python
from collections import Counter
Counter(clustering.labels_).most_common(), len(Counter(clustering.labels_).most_common())
```

```python
data['labels'] = clustering.labels_
```

```python
out_data = data.drop(columns=['embeddings'])
```

```python
out_data.to_csv('outdata.csv')
```

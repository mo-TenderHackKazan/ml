```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
import pandas as pd

df = pd.read_csv('prod.csv')
```

```python
import re
def has_digits(text):
    return bool(re.search("[0-9]", text))

def clean_from_digits(sent):
    words = sent.split()
    new_sent = []
    for word in words:
        if has_digits(word): continue
        new_sent.append(word)
    return ' '.join(new_sent)
```

```python
vectorizer = TfidfVectorizer(max_features=2000)
X = vectorizer.fit_transform(df['filtered'].map(clean_from_digits))
```
